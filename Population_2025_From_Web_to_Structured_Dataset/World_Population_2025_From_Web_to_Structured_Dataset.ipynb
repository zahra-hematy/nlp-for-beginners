```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('span')
```

```python
titles = soup.find_all('span')
title = [title.text.strip() for title in titles]
title
```

```python
filtered_titles = [item for item in title if item not in ['', '#', 'English']]
filtered_titles
```

```python
rows = soup.find_all('tr')[1:]
```

```python
row_draft = [title.text.strip() for title in rows]
row_draft
```

```python
import pandas as pd
df = pd.DataFrame(columns = filtered_titles)
df
```

```python
idx=0
df.insert(loc=idx, column='A', value='new_col')
```

```python

for ro in rows:
    cells = ro.find_all('td')
    row_data = [cell.text.strip() for cell in cells]

    # Adding to the DataFrame
    df.loc[len(df)] = row_data

print(df)
```

```python
df.head(4)
```

```python
df.drop('A', axis=1, inplace=True)
df
```

```python
df.to_csv('Countries in the world by population (2025).csv', index=False)
```
